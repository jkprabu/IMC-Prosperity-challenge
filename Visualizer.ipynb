```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams["figure.figsize"] = [20, 10]
plt.rcParams["figure.autolayout"] = True

df = pd.read_csv("prices_round_1_day_-1.csv", sep = ';')

bdf = df[(df['product'] == 'BANANAS')]

bts = list(bdf['timestamp'])
bp_l = list(bdf['profit_and_loss'])

bmid_pr = list(bdf['mid_price'])

pdf = df[(df['product'] == 'PEARLS')]

pts = list(pdf['timestamp'])
pp_l = list(pdf['profit_and_loss'])

pmid_pr = list(pdf['mid_price'])

plt.figure(figsize=(20,10))
plt.plot(bts,bp_l,linewidth=0.5, c = 'blue')
plt.plot(bts,bmid_pr,linewidth=3, c = 'blue')

plt.plot(pts,pp_l,linewidth=0.5, c = 'red')
plt.plot(pts,pmid_pr,linewidth=3, c = 'red')

plt.title("DAY")
plt.show()
```

```python
df2 = pd.read_csv('prices_round_1_day_0.csv')

b_df = df2[(df2['product'] == 'BANANAS')]

ts = list(b_df['timestamp'])

b_bid1ts = list(b_df['bid_price_1'])
b_bid2ts = list(b_df['bid_price_2'])
b_bid3ts = list(b_df['bid_price_3'])

b_ask1ts = list(b_df['ask_price_1'])
b_ask2ts = list(b_df['ask_price_2'])
b_ask3ts = list(b_df['ask_price_3'])

plt.figure(figsize=(20,10))

plt.scatter(ts,b_bid1ts,linewidth=0.1, c = 'blue')
plt.scatter(ts,b_bid2ts,linewidth=0.1, c = 'blue')
plt.scatter(ts,b_bid3ts,linewidth=0.1, c = 'blue')

plt.show()
```
